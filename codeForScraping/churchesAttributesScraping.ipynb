```python
def getnameList(info):
    Name = info.find("h1",attrs={'id':'firstHeading'}).get_text() if info.find("h1",attrs={'id':'firstHeading'}) else '#'
    nameList.append(Name)
```

```python
def getimageSrcList(info):
    imageSrc = info.find(class_='infobox-image') if info.find(class_='infobox-image') else '#'
    if imageSrc != '#':
        imageSrc = imageSrc.img if imageSrc.img else '#'
    if imageSrc != '#':
        imageSrc = imageSrc['src']
    imageSrcList.append(imageSrc)
```

```python
def getlocationList(info):
    loc = info.find("th",class_="infobox-label",string="Location") if info.find("th",class_="infobox-label",string="Location") else  "no location"
    Location = ''
    if loc == 'no location':
        Location = '#'
    else:
        par = loc.parent if loc.parent else "no parent"
        if par == 'no parent':
            Location = '#'
        else:
            Location = par.td.get_text().replace("\xa0","") if par.td else '#'
    locationList.append(Location)
```

```python
df = pd.read_excel('churchesNamesLinks.xlsx')
mylist = df['Link'].tolist()

for i in mylist:
    url =  i
    data = requests.get(url)
    html = BeautifulSoup(data.text, 'html.parser')
    info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
    if info != 'no infobox':
        getnameList(html)
        getimageSrcList(info)
        getlocationList(info)
        getdistrictList(info)
        getcountryList(info)
        getprovinceList(info)
        gettraditionList(info)
        getdenominationList(info)
        getmembershipList(info)
        getweeklyattendanceList(info)
        getwebsiteList(info)
        getfoundedList(info)
        getfoundersList(info)
        getdedicationList(info)
        getconsecratedList(info)
        geteventsList(info)
        getarchitectsList(info)
        getarchitecturaltypeList(info)
        getstyleList(info)
        getyearsbuiltList(info)
        getgroundbreakingList(info)
        getcompletedList(info)
        getconstructioncostList(info)
        getcapacityList(info)
        getlengthList(info)
        getwidthList(info)
        getheightList(info)
        getchancellorList(info)
        getpriestList(info)
        getministerList(info)
        getpastorList(info)
        getLatitubeLongitudeList(info) 
```
