```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

df = pd.read_excel("etemple_jain_temples.xlsx")
mylist = df['Links'].tolist()

for i in mylist:
    url = i
    data = requests.get(url)
    html = BeautifulSoup(data.text, 'html.parser')
    divtag = html.find('div', class_ = 'section_data active')
    if divtag:
        print(divtag.text)
    else:
        print('nan')
```
