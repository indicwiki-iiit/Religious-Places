```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

def getAddress(html):
    add = html.find('span',class_='text gaddress')
    if add:
        print(add.text)
    else:
        print('nan')

df = pd.read_excel("etemple_jain_temples.xlsx")
mylist = df['Links'].tolist()

for i in range(469, 1100):
    url = mylist[i]
    data = requests.get(url)
    html = BeautifulSoup(data.text, 'html.parser')
    if html:
        getAddress(html)
    else:
        print('nan')
    
```
