```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_excel('only indian wiki links for attributes scraping.xlsx')
mylist = df['links'].tolist()
Name_list = []
Location_list = []
District_list = []
State_list = []
Country_list = []
Province_list = []
Completed_list = []
```

```python
def getLocation():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        Name = ''
        Location = ''
        if info == "no infobox":
            Location = 'nan'
        else:
            Name = info.find("h1",attrs={'id':'firstHeading'}).get_text() if info.find("h1",attrs={'id':'firstHeading'}) else 'nan'
            loc = info.find("th",class_="infobox-label",string="Location") if info.find("th",class_="infobox-label",string="Location") else  "no location"
            if loc == 'no location':
                Location = 'nan'
            else:
                par = loc.parent if loc.parent else "no parent"
                if par == 'no parent':
                    Location = 'nan'
                else:
                    Location = par.td.get_text()
        Location_list.append(Location)
        Name_list.append(Name)
```

```python
def getDistrict():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        District = ''
        if info == "no infobox":
            District = 'nan'
        else:
            dist = info.find("th",class_="infobox-label",string="District") if info.find("th",class_="infobox-label",string="District") else  "no district"
            if dist == 'no district':
                District = 'nan'
            else:
                par = dist.parent if dist.parent else "no parent"
                if par == 'no parent':
                    District = 'nan'
                else:
                    District = par.td.get_text()
        District_list.append(District)
```

```python
def getState():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        State = ''
        if info == "no infobox":
            State = 'nan'
        else:
            st = info.find("th",class_="infobox-label",string="State") if info.find("th",class_="infobox-label",string="State") else  "no state"
            if st == 'no state':
                State = 'nan'
            else:
                par = st.parent if st.parent else "no parent"
                if par == 'no parent':
                    State = 'nan'
                else:
                    State = par.td.get_text()
        State_list.append(State)
```

```python
def getCountry():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        Country = ''
        if info == "no infobox":
            Country = 'nan'
        else:
            ctry = info.find("th",class_="infobox-label",string="Country") if info.find("th",class_="infobox-label",string="Country") else  "no country"
            if ctry == 'no country':
                Country = 'nan'
            else:
                par = ctry.parent if ctry.parent else "no parent"
                if par == 'no parent':
                    Country = 'nan'
                else:
                    Country = par.td.get_text()
        Country_list.append(Country)
```

```python
def getProvince():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        Province = ''
        if info == "no infobox":
            Province = 'nan'
        else:
            pro = info.find("th",class_="infobox-label",string="Province") if info.find("th",class_="infobox-label",string="Province") else  "no Province"
            if pro == 'no Province':
                Province = 'nan'
            else:
                par = pro.parent if pro.parent else "no parent"
                if par == 'no parent':
                    Province = 'nan'
                else:
                    Province = par.td.get_text()
        Province_list.append(Province)
```

```python
def getCompleted():
    for i in mylist:
        url = i
        data = requests.get(url)
        html = BeautifulSoup(data.text, 'html.parser')

        info = html.find("table", class_="infobox") if html.find("table", class_="infobox") else "no infobox"
        Completed = ''
        if info == "no infobox":
            Completed = 'nan'
        else:
            list_to_find = ['Completed','Date established','Date', 'Groundbreaking']
            i = 0
            flag = 0
            while i < 4:
                if info.find("th",class_="infobox-label",string=list_to_find[i]):
                    comp = info.find("th",class_="infobox-label",string=list_to_find[i]) 
                    flag = 1
                    break
                i += 1
            if flag == 0:
                Completed = 'nan'
            else:
                par = comp.parent if comp.parent else "no parent"
                if par == 'no parent':
                    Completed = 'nan'
                else:
                    Completed = par.td.get_text()
        
        Completed_list.append(Completed)
    
```

```python
getLocation()
getDistrict()
getState()
getCountry()
getProvince()
getCompleted()
```

```python
mosque_attrs = pd.DataFrame({ "MOSQUE NAME": Name_list ,"Location": Location_list,"District":District_list,"state":State_list,"Country":Country_list,'Province':Province_list,'Completed':Completed_list})   
mosque_attrs.to_excel("only india  attributes scaraped.xlsx")
```
